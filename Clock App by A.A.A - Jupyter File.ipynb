```python
#The Clock App - Pyhton Microproject - By Agnivesh PS (8), Amith Shaji George (11), Athul S (21) - S2 CS1

import tkinter as tk

#For using time related functions
from time import strftime

#For VS Code Stability
import time 

#To play sound
import pygame

#To display pop-up window
from tkinter import messagebox

# Function to update the clock
def update_clock():
    current_time = time.strftime("%H:%M:%S")
    clock_label.config(text=current_time)
    root.after(1000, update_clock)  # Update the clock every 1000 milliseconds (1 second)

#Function for Stopwatch
def stopwatch():
    def start_stopwatch():
    
        #Here elaspsed_time and running act as non local variable for start_stopwatch
        nonlocal elapsed_time, running
        if not running:
            running = True
            update_stopwatch()

    def stop_stopwatch():

        #Here running ac as non local variable for stop_stopwatch
        nonlocal running
        running = False

    def reset_stopwatch():
        nonlocal elapsed_time, running
        running = False
        elapsed_time = 0
        stopwatch_label.config(text="00:00")

    def update_stopwatch():

        #Here elaspsed_time and running act as non local variable for update_stopwatch
        nonlocal elapsed_time, running
        if running:
            elapsed_time += 1
            minutes = elapsed_time // 60
            seconds = elapsed_time % 60

            #02 are used to intake 2 digit time variable
            stopwatch_label.config(text=f"{minutes:02}:{seconds:02}",font=('Gilroy-Bold',25))

            #The update_stopwatch function is called after each 1000 ms (1 sec)
            root.after(1000, update_stopwatch)

    root = tk.Tk()
    root.title("Stopwatch")
    root.geometry('300x280')
    root.configure(bg='#171717')

    elapsed_time = 0
    running = False

    stopwatch_label = tk.Label(root, text="00:00", font=('Gilroy-Bold',25),fg='white', bg='#171717')
    stopwatch_label.pack(pady=20)

    start_button = tk.Button(root, text="Start", command=start_stopwatch, width='12', font=('Gilroy-Bold',12))
    start_button.pack(pady=5)

    stop_button = tk.Button(root, text="Stop", command=stop_stopwatch, width='12', font=('Gilroy-Bold',12))
    stop_button.pack(pady=5)

    reset_button = tk.Button(root, text="Reset", command=reset_stopwatch, width='12', font=('Gilroy-Bold',12))
    reset_button.pack(pady=5)

#Function for Timer

def timer_function():
    timer_window = tk.Toplevel(root)
    timer_window.title("Timer")
    timer_window.geometry('300x280')
    timer_window.configure(bg='#171717')

    #StringVar is used to specify a pre-defined data for an entry
    time_var = tk.StringVar()
    time_var.set("00:00:00")
    
    def start_timer():
        try:
            #Map is used as an alternative for iteration statements
            h, m, s = map(int, time_var.get().split(":"))
            time_left = h * 3600 + m * 60 + s
        except ValueError:
            messagebox.showerror("Invalid Time", "Please enter time in HH:MM:SS format")
            return
        
        def countdown():
            nonlocal time_left
            if time_left > 0:
                time_left -= 1
                h, m, s = time_left // 3600, (time_left % 3600) // 60, time_left % 60
                time_var.set(f"{h:02}:{m:02}:{s:02}")

                #countdown function is repeated after every 1000 ms (1 sec)
                timer_window.after(1000, countdown)
            elif time_left == 0:
                messagebox.showinfo("Time's up", "The countdown has finished!")
        
        countdown()
    
    def reset_timer():
        time_var.set("00:00:00")
    
    time_entry = tk.Entry(timer_window, textvariable=time_var, width='12',font=('Gilroy-Bold',20),justify='center')
    time_entry.pack(pady=20)
    
    start_button = tk.Button(timer_window, text="Start", command=start_timer,width='12', font=('Gilroy-Bold',12))
    start_button.pack(padx=20, pady=20)
    
    reset_button = tk.Button(timer_window, text="Reset", command=reset_timer, width='12', font=('Gilroy-Bold',12))
    reset_button.pack(padx=20)

#Function for Alarm
#pygame.mixer.init() is used to load and play songs

pygame.mixer.init()

def set_alarm():

    #get funtion is used to get the data entered in the entry column
    alarm_time = f"{hour_var.get()}:{minute_var.get()}:{second_var.get()}"
    alarm_label.config(text=f"Alarm set for: {alarm_time}")
        
    def check_alarm():
        current_time = time.strftime("%H:%M:%S")
        if current_time==alarm_time:
            #load is used to load a music file for a directory
            pygame.mixer.music.load("alarmclock.mp3")
            
            #play is used to play a music
            pygame.mixer.music.play()

            #messagebox.showinfo is used to display contents in an external pop-up window
            messagebox.showinfo("Alarm", "Time to wake up!")
            return
        
        #Calls the check_alarm function after each second
        alarm_window.after(1000, check_alarm)
    check_alarm()

def open_alarm_window():
    global alarm_window
    alarm_window = tk.Toplevel(root)
    alarm_window.title("Alarm Clock")
    alarm_window.geometry('300x210')
    alarm_window.configure(bg='#171717')
        
    global hour_var, minute_var, second_var, alarm_label
        
    # Create variables to store alarm time
    hour_var = tk.StringVar(value="00")
    minute_var = tk.StringVar(value="00")
    second_var = tk.StringVar(value="00")
        
    # Create and place widgets
    tk.Label(alarm_window, text="Set Alarm Time (24-hour format)", font=('Gilroy-Bold',10),bg='#171717',fg='white').pack(pady=10)
        
    frame = tk.Frame(alarm_window)
    frame.pack()
        
    tk.Entry(frame, textvariable=hour_var, width=3, font=('Gilroy-Bold',20), justify='center').pack(side=tk.LEFT)
    tk.Label(frame, text=":", font=('Gilroy-Bold',20)).pack(side=tk.LEFT)
    tk.Entry(frame, textvariable=minute_var, width=3, font=('Gilroy-Bold',20), justify='center').pack(side=tk.LEFT)
    tk.Label(frame, text=":", font=('Gilroy-Bold',20)).pack(side=tk.LEFT)
    tk.Entry(frame, textvariable=second_var, width=3, font=('Gilroy-Bold',20), justify='center').pack(side=tk.LEFT)
        
    tk.Button(alarm_window, text="Set Alarm", command=set_alarm, font=("Gilroy-Bold",10)).pack(pady=20)
        
    alarm_label = tk.Label(alarm_window, text="", font=('Gilroy-Bold',15), bg='#171717', fg='white')
    alarm_label.pack()

    root.mainloop()


# Initialize the main window
root = tk.Tk()
root.title("The Clock App")
root.geometry('300x280')
root.configure(bg='#171717')

#To create a frame which keeps the elements in place (Responsive Page)
frame = tk.Frame(root)
frame.pack()

# Create and place the clock label
clock_label = tk.Label(root, font=("Gilroy-Black",30), background='#171717', foreground='white')
clock_label.pack(pady=20)

# Create and place the buttons
button1 = tk.Button(root, text="Timer", width='12', font=('Gilroy-Bold',12), command=timer_function)
button1.pack(pady=10)

button2 = tk.Button(root, text="Stopwatch", width='12', font=('Gilroy-Bold',12), command=stopwatch)
button2.pack(pady=10)

button3 = tk.Button(root, text="Alarm", width='12', font=('Gilroy-Bold',12), command=open_alarm_window)
button3.pack(pady=10)

# To start the live clock
update_clock()

# To enable home screen
root.mainloop()
```
